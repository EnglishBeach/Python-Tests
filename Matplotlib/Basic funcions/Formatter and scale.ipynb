```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator,FixedLocator
from matplotlib.ticker import NullFormatter,FormatStrFormatter,FuncFormatter,ScalarFormatter,FixedFormatter
```

```python
x= np.linspace(-10,10,100)
```

```python
fig = plt.figure()
fig.set(facecolor='yellow', figheight=8, figwidth=10)
gs = plt.GridSpec(2, 2, figure=fig)

ax1 = plt.subplot(gs[0:2, 0])
# ax1.plot(x, x**3 / 5)
ax1.plot(x,x)
ax1.xaxis.set_major_formatter(NullFormatter())
scal_form = ScalarFormatter(10)
scal_form.set_powerlimits((-1, 1))
ax1.yaxis.set_major_formatter(scal_form)
# ax1.set_yscale('log',base=10)
ax1.set_xscale('symlog',base=2,linthresh=3,linscale=5)

def format_func(x, pos):
    return f'label {x:.2}'


ax2 = plt.subplot(gs[0, 1])
ax2.scatter(np.arange(20), np.random.random(20), c='green')
ax2.yaxis.set_minor_locator(MultipleLocator(0.1))
ax2.yaxis.set_minor_formatter(FuncFormatter(format_func))
ax2.xaxis.set_major_formatter(FormatStrFormatter('x= %.2f'))


ax3 = plt.subplot(gs[1, 1])
ax3.plot(x, (np.sinc(x) - 1)**2)
ax3.xaxis.set_major_locator(FixedLocator([-10, -7, -5, 0, 1, 2, 100]))
ax3.xaxis.set_major_formatter(FixedFormatter(['a -10', 'b', 'c', 'd', '1',' 2', 'finish']))
ax3.minorticks_on()
ax3.grid(which='major',color='black',linewidth=1)
ax3.grid(which='minor',color='red',ls=':')
```
