```python
from scipy.integrate import odeint,solve_ivp
import matplotlib.pyplot as plt
import numpy as np
```

```python
import sympy as sm
y1,y2 = sm.symbols('y1 y2')
```

```python
def f(t, Y):
    y1, y2 = Y
    A = np.array([[3, -4], [1, -2]])
    return A @ Y


Y0 = [3, 2]
```

```python
t = np.linspace(0,30,300)
Y_sol = odeint(func=f,t=t,tfirst=True,y0=Y0)
```

```python
y1_sol ,  y2_sol = Y_sol[:,0], Y_sol[:,1];
plt.plot(t,y1_sol)
plt.plot(t,y2_sol)
```
