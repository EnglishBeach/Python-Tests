### Sorting & grouping ...

```python
df.sort_values(by=['Pclass','Age'],ascending=[True,False])
```

```python
mean_m = df[df['Sex'] == 'male']['Age'].mean()
mean_m = round(mean_m, 3)

mean_f = df[df['Sex'] == 'female']['Age'].mean()
mean_f = round(mean_f, 3)
print(df['Age'].mean())
print(mean_m, mean_f, sep=sepr)
```

```python
df.groupby(by='Survived')[[ 'Age']].describe()
```

```python
pd.crosstab(df['Age'],df['Survived'],margins=True)
```

```python
print(
df.pivot_table(['Age','Fare'],['Survived']),
df.groupby(by='Survived')[['Age','Fare']].mean(),
sep=sepr
)
```

```python
print(
p_df := df.pivot_table(['Age','Fare'],['Survived','Sex','Pclass'],aggfunc='max'),
df.groupby(by=['Survived','Sex','Pclass'])[['Age','Fare']].max(),
sep=sepr
)
```

### Indexing

```python
n = 2
print(
    df.loc[0:n]['Sex'],
    df.loc[0:n, 'Sex'],
    df['Sex'].loc[0:n],
     # iloc
    df.iloc[0:n + 1, 3],
     # slice
    df[0:n + 1]['Sex'],
    df['Sex'][0:n + 1],
    sep=sepr,
)
```

```python
df.iloc[0:4]
n = 2
print(
    # df.loc[0:n]['Sex':'Age'],
    df.loc[0:n, 'Sex':'Age'],
    df.iloc[0:n + 1, 3:5],
    df[['Sex','Age']][0:n+1],
    df.loc[[1,2], 'Sex':'Age'],
    sep=sepr,
)
```
